```python
y = df['quality']
X = df.drop(columns=['quality', 'fixed acidity', 'volatile acidity', 'citric acid', 'chlorides', 'sulphates', 'alcohol', 'pH'], axis=1)
y.shape
X.shape
```

```python
pip install statsmodels
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calcu_mult(df):
  vif = pd.DataFrame()
  vif["variables"] = X.columns
  vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

  return vif
```

```python
calcu_mult(df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=45)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
std_scale = MinMaxScaler() 
X_train_std = pd.DataFrame(std_scale.fit_transform(X_train))
X_train_std
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1, max_depth=4)
model.fit(df,X)

RandomForestRegressor(max_depth=4, random_state=1)
```

```python
features = df.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-3:]
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='green', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
model = LinearRegression()
w_quality = model.fit(X_val, y_val)
w_quality
```

```python
coef_deter = w_quality.score(X_val, y_val)
coef_deter
```

```python
1 - (1-model.score(X_val, y_val))*(len(y_val)-1)/(len(y_val)-X_val.shape[1]-1)
```

```python
y_predict = w_quality.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("MAE : ",mean_absolute_error(y_test, y_predict))
print("MSE : ",mean_squared_error(y_test, y_predict))
print("R2 score : ",r2_score(y_test, y_predict))
```
